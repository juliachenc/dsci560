```python
df = pd.read_csv("/content/drive/MyDrive/Fall 2021 DSCI560/data_sepTime/29.txt") 
df.columns = ['building', 'location', 'ClientMacAddr','latitude','longitude','date','time']
```

```python
workHR = df[(df['time'] > "06:00:00") & (df['time'] < "22:00:00")] 
```

```python
count_addr_bydate = workHR.groupby(by = ["ClientMacAddr", "date"]).size().reset_index()
count_addr_bydate.columns = ['ClientMacAddr', 'date','count']
count_addr_bydate.sort_values(by = 'count', inplace=True)
```

```python
# p = count_addr_bydate.plot(kind='hist', color='r', alpha=0.5)
```

```python
todrop = count_addr_bydate[(count_addr_bydate['count']>3000) | (count_addr_bydate['count'] < 3)]
```

```python
listAdd = todrop["ClientMacAddr"].to_list()
```

```python
df_idx = workHR[workHR["ClientMacAddr"].isin(listAdd)].index
newdf = workHR.drop(df_idx)
```

```python
count = df.groupby(by = "date").size().reset_index()
count.columns = ['date','count']
count.sort_values(by = 'date', inplace=True)
```

```python
newdf.to_csv('29.txt', header=True, index=False, sep=',')
count.to_csv('count_29.txt', header=True, index=False, sep=',')
```
